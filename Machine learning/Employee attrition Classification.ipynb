### Data Preprocessing

#### Convert Attrition from ('Yes', 'No') to (1,0) 

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
df.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

```python
newdf = pd.get_dummies(df, columns=['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime'])
newdf.head()
```

#### Check the shape of our new dataset

```python
newdf.shape
```

#### Print unique values in our new dataframe

```python
for column in newdf.columns:
    print(f"Unique values in column '{column}' are: {newdf[column].unique()}")
```

#### Split the columns into input and target variables (X and y)

```python
X = df.drop(columns=['Attrition'])
y = df['Attrition']
```

#### Carry out Feature scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = pd.get_dummies(X)
```

```python
X  = pd.DataFrame(scaler.fit_transform(X),columns = X.columns)
```

```python
X.head()
```

### Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size = 0.25)
```

### Random Forest Classifier
** Choose the best estimator and parameters :GridSearchCV**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size = 0.25)
```

```python
params = [{'n_estimators':list(range(50,60)),
           'max_depth': [1,10],
           }]

```

```python
gsv = GridSearchCV(RandomForestClassifier(), cv=5, param_grid=params, scoring='accuracy')
```

```python
gsv.fit(X_train,y_train)
```

```python
y_pred = gsv.predict(X_test)

score = gsv.score(X_test, y_test)
print("Score: ", score)
```

<span style = "color:blue"> Create Random forest model with the best parameters </span>

```python
best_estimator = RandomForestClassifier(max_depth=10, n_estimators=32, n_jobs=1, random_state=20)
                                        
best_estimator.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score
```

```python
y_pred = best_estimator.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)
cv_score = cross_val_score(best_estimator, X, y, cv=5)
print("Accuracy score: ", accuracy)
print("Confusion matrix: \n", cm)
print("Classification report: \n", cr)
print("Cross val score: ", cv_score.mean())
```

<span style = "color:orange"> Visualize confusion matrix </span>

```python
sns.heatmap(metrics.confusion_matrix(y_test, y_pred), annot = True,fmt = 'd')
plt.show()
```
